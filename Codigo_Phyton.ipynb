```python
import time
import matplotlib.pyplot as plt


def factorial_iterativo(n):
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado


def factorial_recursivo(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursivo(n - 1)


valores_n = list(range(1, 21))
tiempos_iter = []
tiempos_rec = []

for n in valores_n:
    inicio = time.time()
    factorial_iterativo(n)
    tiempos_iter.append(time.time() - inicio)

    inicio = time.time()
    factorial_recursivo(n)
    tiempos_rec.append(time.time() - inicio)


plt.plot(valores_n, tiempos_iter, marker="o", label="Iterativo")
plt.plot(valores_n, tiempos_rec, marker="s", label="Recursivo")
plt.xlabel("n")
plt.ylabel("Tiempo (s)")
plt.title("Comparación Factorial Iterativo vs Recursivo")
plt.legend()
plt.grid(True)
plt.savefig("grafico.png", dpi=300)
plt.show()

```
